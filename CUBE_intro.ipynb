#### Чтобы начать работать с большими данными нужно уметь "запаковывать информацию", в математике для этого используют вектора.
> **Вектором** называется направленный отрезок, для которого указано его начало и конец

  vektory.jpg
  
> В данном случае началом отрезка является точка *A* концом отрезка – точка *B*. Сам вектор обозначается ̅AB̅. Направление имеет существенное значение, если переставить стрелку в другой конец отрезка, то получится вектор B̅A̅ и это уже **совершенно другой вектор**! Понятие вектора удобно отождествлять с движением физического тела: согласитесь, зайти в двери института или выйти из дверей института – это совершенно разные вещи.
Дополнительную информацию можно найти на [сайте](http://mathprofi.ru)

#### Скалярное произведение векторов
image.png

> По определению скалярное произведение вектов - это произведение длин векторов на косинус угла между ними. Этот подход является общим иии скучным, вряд ли в реальной жизни вы будете знать и длину векторов, и косинус угла между ними. 

```python
mul1 = 5 * 12 * np.cos(1)
mul2 = 2 * 3 * np.cos(0)
mul1, mul2
```

> Нам же интересно произведение векторов, у которых нам известны только координаты

```python
x = np.array([1,3,5])
y = np.array([4, 7, -2])
dot = x * y  #мы нашли произведение координат, чтобы найти скалярное произведение
             #нужно суммировать эти значения!
dot.sum()
```
